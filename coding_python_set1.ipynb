```python
def factorial(n):
    result=1
    for i in range(1,n+1):
        result*=i
    return result
print(factorial(5))
```

```python
num=int(input("Enter the number:"))
if num<=1:
    print("Not a prime number")
else:
    for i in range(2,num):
        if num%i==0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```
